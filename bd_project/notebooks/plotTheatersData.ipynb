```python
import pickle
outA=pickle.load(open('../scripts/Apollo_24h_7d.out','rb'))
out=pickle.load(open('../scripts/all_24h_7d.out','rb'))
outL=pickle.load(open('../scripts/Lincoln_24h_7d.out','rb'))
out5=pickle.load(open('../scripts/59E59_24h_7d.out','rb'))
outAT = pickle.load(open('../scripts/allTheaters_24h_7d.out','rb'))
```
