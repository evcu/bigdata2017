```python
import pickle
out1=pickle.load(open('../scripts/lon0.002000_lat0.001200.out','rb'))
out2=pickle.load(open('../scripts/lon0.004000_lat0.002400.out','rb'))
out3=pickle.load(open('../scripts/lon0.008000_lat0.005000.out','rb'))
out4=pickle.load(open('../scripts/lon0.016000_lat0.010000.out','rb'))
outall=pickle.load(open('../scripts/barclays_all.out','rb'))
outs = [out1,out2,out3,out4,outall]
labels = ['Box1','Box2','Box3','Box4','All-Data']
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

for out,lab in zip(outs,labels):
    x=[e[0][2] for e in out]
    y=np.array([e[1] for e in out])
    y=y/np.mean(y)
    plt.scatter(x,y, label=lab)
#     x=[e[0][2] for e in out if e[0][1]==i]
#     y=np.array([e[1] for e in outL if e[0][1]==i])
#     y=y/np.mean(y)
#     plt.scatter(x,y, label='The Walter Reade Theater at Lincoln Center')
#     x=[e[0][2] for e in out if e[0][1]==i]
#     y=np.array([e[1] for e in out5 if e[0][1]==i])
#     y=y/np.mean(y)
#     plt.scatter(x,y, label='59E59')
#     x=[e[0][2] for e in outA if e[0][1]==i]
#     y=np.array([e[1] for e in outA if e[0][1]==i])
#     y=y/np.mean(y)
#     plt.scatter(x,y, label='Apollo Theater')

plt.title('Hourly pickup counts around Barclays at event dates.')
plt.ylabel('Normalized taxi pickups')
plt.xlabel('Hour of the day')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2),
      fancybox=True, shadow=True, ncol=5)
plt.show()
```
